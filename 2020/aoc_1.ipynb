```python
#aoc 2020 day 1
Exp=['']
with open('input.txt') as f:    
    for i in f:
        Exp.append(int(i))
    Exp.pop(0)
    print(len(Exp))
    print(Exp)
        
    for j in range(0,len(Exp)):
        a=Exp[j]
        for k in range(1,len(Exp)):
            b=Exp[k]
            
            if a+b == 2020:
                c = a*b
                print("2 numbers: ",a,b,a+b,c)
                
            for l in range(2,len(Exp)):
                d=Exp[l]
                
                if a+b+d == 2020:
                    e = a*b*d
                    print("3 numbers: ",a,b,d,a+b+d,e)
```
